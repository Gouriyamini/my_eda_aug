- If you want to drop a specific column

- we will use drop method

- we already know about difference between keywords and methods

- keywords are used directly

- Methods are used with correspondind dataframe

- **Method name: drop**

df.drop()

- column: which column you want to drop
    
- axis: Represents the specified value is row or column
    
    - axis=0 means rows
    
    - axis=1 means column
    
    - by default it takes axis=0 

- inplace: will provide overwrite the dataframe or nor
   
   - inplace = true, this indicates overwriting the dataframe
   
   - inplace = False, this indicates the modified dataframe will be used when you provide new variable for the operation   

```python
list2=[i**3 for i in range(1,11)]
df['new column']=list2
df
```

```python
df.drop('new column',axis=1,inplace=True)
df
```

**step-12 how to avoid index column**

- while we save the dataframe

- automatically a new column will be created the we need to avoid

```python
# avoiding the extra or index column
# don't perform this while file is opened 
# other wise error permission denied occurs

df2.to_csv('Data1.csv',index=False)
df2.to_excel('Data1.xlsx',index=False)
```
